# Churn Prediction

In einer allgemeineren Definition ist *Churn* eine Kennzahl, die die Anzahl der Kunden angibt, die  in einem bestimmten Zeitraum kündigen . Um die Abwanderung zu berechnen, müssen Sie lediglich die Anzahl der Kunden addieren, die Ihr Produkt/Ihre Dienstleistung im analysierten Zeitraum gekündigt haben.

//
<p align=center>
<img src="https://vanessamiranda.github.io/images/Customerchurn/customerchurn.jpg" width="60%"></p>



Die Abwanderungsrate wird im Allgemeinen über einen bestimmten Zeitraum berechnet, sei es jährlich, halbjährlich oder monatlich. Beispiel: Wenn jeden Monat einer von 20 Kunden Ihr Produkt kündigt, bedeutet dies, dass die Abwanderungsrate für Ihr Produkt 5 % beträgt.
Sie können auch die Abwanderungsrate, die den Prozentsatz der Konten darstellt, die gekündigt werden, im Vergleich zum aktiven Kundenstamm wie folgt berechnen:
<br>


\begin{align}
Abwanderungsrate  = \frac{Gesamtzahl\ gekuendigter\ Kunden}{Gesamtzahl \ aktiver\ Kunden}
\end{align}

<br>
<br>

Die Kundenabwanderung wirkt sich auf die Rentabilität aus, besonders in Branchen, in denen die Einnahmen stark von Abonnements abhängig sind (z. B. Banken, Telefon- und Internetdienstanbieter, Pay-TV-Unternehmen, Versicherungsunternehmen usw.). Es wird geschätzt, dass die Gewinnung eines neuen Kunden bis zu fünfmal mehr kosten kann, als die Bindung eines bestehenden Kunden. Daher ist die Kundenabwanderungsanalyse unerlässlich für ein Unternehmen. Das Ziel dieses Projektes ist es, die Churn für eine Unternehmen zu verstehen und vorherzusagen.

Das Ziel der Churn Prediction ist es, diese abwanderungsgefährdeten Kunden zu identifizieren, bevor sie das Unternehmen verlassen, und präventive Maßnahmen zu ergreifen, um sie zu halten. Dies kann bedeutsam sein, da die Neukundengewinnung oft teurer ist als die Bindung bestehender Kunden.

### Die Herausforderung der Analyse

Eine der Haupt Herausforderungen besteht darin, genaue Vorhersagen darüber zu treffen, welche Kunden wahrscheinlich abwandern werden. Dies erfordert eine gründliche Analyse einer Vielzahl von Daten, die das Kundenverhalten und die Interaktionen mit dem Unternehmen beschreiben. Jedoch können Datenqualität und -quantität eine Herausforderung darstellen, da Unternehmen möglicherweise nicht über ausreichende oder qualitativ hochwertige Daten verfügen, um präzise Vorhersagen zu treffen.

Ein weiteres Problem ist die Bewältigung von unbalancierten Daten, da die Anzahl der abwandernden Kunden oft viel kleiner ist als die Anzahl der nicht abwandernden Kunden. Dies kann zu Verzerrungen führen und die Leistung des Modells beeinträchtigen. Es ist wichtig, Techniken zur Bewältigung von unbalancierten Daten anzuwenden, um genaue Vorhersagen zu gewährleisten.

Zusätzlich können komplexe Kundenverhaltensmuster und Wechselwirkungen zwischen verschiedenen Merkmalen eine Herausforderung darstellen. Das Identifizieren relevanter Merkmale und das Verständnis ihrer Auswirkungen auf das Churn-Verhalten erfordert oft Fachkenntnisse und fortgeschrittene Analysetechniken.

Die Vorhersagegenauigkeit ist ein weiteres Anliegen, da Vorhersagemodelle oft Overfitting oder Underfitting zeigen können. Overfitting tritt auf, wenn das Modell zu stark an die Trainingsdaten angepasst ist und nicht in der Lage ist, auf neuen Daten gut zu generalisieren. Underfitting hingegen tritt auf, wenn das Modell zu einfach ist und nicht in der Lage ist, die zugrunde liegenden Muster der Daten zu erfassen.

Datenschutz und Datenschutz sind ebenfalls wichtige Aspekte bei der Churn-Analyse, da sie den Umgang mit sensiblen Kundeninformationen betreffen. Es ist entscheidend sicherzustellen, dass alle Datenschutzbestimmungen eingehalten werden und dass Kundeninformationen sicher und vertraulich behandelt werden.

Trotz dieser Herausforderungen bietet die Churn-Analyse Unternehmen die Möglichkeit, Kundenbeziehungen zu stärken und langfristige Kundenloyalität aufzubauen. Durch den Einsatz fortgeschrittener Analysetechniken und die Berücksichtigung von Datenschutzbestimmungen können Unternehmen präzise Vorhersagen treffen und wirksame Strategien zur Kundenbindung entwickeln.

# Se­lek­ti­on kategorialer und numerischer Merkmale

-  **SENIORCITIZEN** ein kategorischer variabler Typ ist und numerisch angezeigt wird.
- Daher müssen wir ihn zunächst in eine kategoriale Funktion umwandeln.
- **TOTALCHARGES** wird als kategorischer Typ angezeigt, aber ist ein numerischer variabler Typ.

```python
# Allen Spalten von der Typ object float und int werden in zwei Listen eingesammelt.
# Die Funktion  summarize_cat gibt die eindeutigen Werte von jeden Spalten zurück, je nach Datentyp in einem Dataframe
# TOTALCHARGES wird als kategorische variablen Typ angezeigt, aber ist ein numerischer Wert.
# Daher müssen wir ihn zunächst in eine numerische umwandeln.

# drop customerID
df.drop("customerID", axis=1, inplace=True)

# Select Spalten von Type object
categorical_features = df.select_dtypes(include=[object]).columns

# TotalCharges wird  als numerische Merkmal behandeln, also wird von Katerorische Merkmalen entfernt und SeniorCitizen eingefügt
categorical_features= list(categorical_features.difference(['TotalCharges'])) + ['SeniorCitizen']

# SeniorCitizen wird  als kategoriales Merkmal behandeln, so das von numerischen Merkmalen entfernt ist und TotalCharge eingefügt.
numerical_features = list(df.select_dtypes(include=[np.float64,np.int64]).columns.difference(['SeniorCitizen'])) + ['TotalCharges']
```

```python
#  Funktion zum Zusammenfassen kategorialer und Numerische Daten durch Auflisten eindeutiger Werte für jede Kategorie in einem Datenrahmen
def summarize_cat(data,categorical_features):
    results=[]

    for column in data[categorical_features]:
        # eindeutigen Werte der Spalte
        members = data[column].unique().tolist()
        # Hängen Sie den Spaltennamen und seine eindeutigen Werte an die liste an
        results.append([column, members])

    return pd.DataFrame(results, columns=['Column Name', 'Members'])

# Das Datenset wurde numerisch und kategorisch eingeteilt. So hat man einen besseren Überblick, wie die Daten verteilt sind.
```

# Spalten transformation
Die Spalten **OnlineSecurity,	OnlineBackup,	DeviceProtection,	TechSupport,**	**StreamingTV,	StreamingMovies**, sind Categorische Merkmale mit dem Werte **['No' 'Yes' 'No internet service','No phone service']**. Hier wird vermutet ein dass Fehler bei der Eintragung in den Datensatz liegt. 'No', 'No phone service' und 'No internet service' haben den gleichen Beteutung.

```python
# Die Funktion transform wandeltet die Werte 'No internet service','No phone service' in No
def transform_no (wert):
  if (wert == 'No phone service' or wert=='No internet service' or wert=='No'):
    return 'No'

  else:
    return'Yes'

df['MultipleLines'] = df['MultipleLines'].map(transform_no)
df['OnlineSecurity'] = df['OnlineSecurity'].map(transform_no)
df['OnlineBackup'] = df['OnlineBackup'].map(transform_no)
df['DeviceProtection'] = df['DeviceProtection'].map(transform_no)
df['TechSupport'] = df['TechSupport'].map(transform_no)
df['StreamingTV'] = df['StreamingTV'].map(transform_no)
df['StreamingMovies'] = df['StreamingMovies'].map(transform_no)

```

Die Spalte **TotalCharges** könnte nicht zum mumerische Spalte ungewandelt werden.
1. df["TotalCharges"].astype('float32')
2. ValueError: could not convert string to float: ' '

info: https://saturncloud.io/blog/how-to-handle-the-pandas-valueerror-could-not-convert-string-to-float/

- Beim Konvertieren der TotalCharges in Float ist ein Fehler mit der Meldung aufgetreten, dass die Zeichenfolge nicht in Float konvertiert werden konnte.
- Diese Meldung wurde angezeigt, weil an den oben angezeigten Indexpositionen der Spalte „TotalCharges“ leere Zeichenfolgen vorhanden waren.
- Da diese Elemente als Zeichenfolgen definiert wurden, wurden sie nicht als Nullwerte angezeigt und daher wurde in der missingno für fehlende Werte nichts angezeigt.
- Beispiel: a = ' '
- Daher teilen wir die einzelnen Elemente von TotalCharges auf und speichern die Indexwerte der Elemente, deren geteilte Stringlänge ungleich 1 ist.
- Dadurch wird die Spalte mit dem Medianwert gefüllt, und schließlich wird die gesamte Spalte mithilfe der Funktion to_numeric in Float umgewandelt.
- Wir streichen auch die Spalte „customerID“!
Unterteilen wir die Features in numerische und kategoriale Features.
Wir werden auch die Label-Codierungstransformation für kategoriale Features durchführen.

```python
# Indexposition von ValueError: could not convert string to float: ' ' ermittelten.Split gibt 0 zurück bei leer Zeichen
l1 = [len(i.split()) for i in df['TotalCharges']]
l2 = [i for i in range(len(l1)) if l1[i] != 1]
print('Index Positionen mit empty spaces : ',l2)

# a = len('hallo'.split())
# a

# Konvertiert die Spalte 'TOTALCHARGES' in numerisch, entfernt Leerzeichen, ersetzt Fehler durch NaN -> Datenbereinigung.
df.TotalCharges = pd.to_numeric(df.TotalCharges.str.strip(), errors='coerce',)
```

# Vertrage und Bezahlungsform

Der Telco Customer Churn Datensatz enthält verschiedene Attribute, die helfen, das Kundenverhalten zu analysieren und die Wahrscheinlichkeit der Kundenabwanderung (Churn) zu bestimmen. Die Spalten **"Contract", "PaperlessBilling" und "PaymentMethod"** spielen hierbei eine entscheidende Rolle. Sie geben Einblick in die Art und Weise, wie Kunden mit dem Unternehmen interagieren und ihre Rechnungen begleichen. Hier sind die Bedeutungen und die Implikationen dieser Spalten.


**1. Contract**
Die Spalte "Contract" gibt die Vertragsart des Kunden an. Es gibt typischerweise drei Vertragsarten: Monatlich, Ein-Jahres-Vertrag und Zwei-Jahres-Vertrag.

- **Monatlich:** Kunden mit einem monatlichen Vertrag haben die größte Flexibilität und können ihren Dienst jederzeit kündigen. Diese Kunden haben eine höhere Wahrscheinlichkeit zu kündigen, da sie weniger vertraglich gebunden sind.

- **Ein-Jahres-Vertrag:** Diese Kunden sind für ein Jahr vertraglich gebunden. Sie haben eine mittlere Bindung und zeigen eine moderate Wahrscheinlichkeit zu kündigen.

- **Zwei-Jahres-Vertrag:** Kunden mit einem Zwei-Jahres-Vertrag sind langfristig an das Unternehmen gebunden. Diese Kunden haben eine geringere Wahrscheinlichkeit zu kündigen, da sie durch den Vertrag länger an das Unternehmen gebunden sind.
Die Vertragsart ist ein starker Indikator für die Kundenbindung und hilft dem Unternehmen, Strategien zur Reduzierung der Kündigungsrate zu entwickeln.

**2. PaperlessBilling**
Die Spalte "PaperlessBilling" gibt an, ob der Kunde die papierlose Abrechnung gewählt hat. Dies bedeutet, dass der Kunde seine Rechnungen elektronisch erhält und bezahlt.

- **Ja:** Kunden, die papierlose Abrechnung nutzen, bevorzugen möglicherweise bequemere und umweltfreundlichere Lösungen. Diese Kunden sind oft
technikaffiner und können höhere Erwartungen an digitale Dienstleistungen haben.

- **Nein:** Kunden, die keine papierlose Abrechnung nutzen, bevorzugen möglicherweise traditionelle Methoden und könnten weniger technikaffin sein.
Das Verständnis der Präferenz für papierlose Abrechnung kann dem Unternehmen helfen, gezielte Marketingstrategien und Serviceverbesserungen zu entwickeln, um die Zufriedenheit und Bindung dieser Kunden zu erhöhen.

**3. PaymentMethod**

Die Spalte "PaymentMethod" gibt die bevorzugte Zahlungsmethode des Kunden an. Es gibt verschiedene Zahlungsmethoden wie Kreditkarte, Banküberweisung, elektronische Schecks und Papier-Schecks.

- **Kreditkarte (automatisch):** Kunden, die automatische Kreditkartenzahlungen nutzen, bevorzugen bequeme und automatisierte Zahlungsmethoden. Diese Kunden haben möglicherweise eine geringere Kündigungsrate, da ihre Zahlungen automatisch abgewickelt werden.

- **Banküberweisung (automatisch):** Ähnlich wie bei Kreditkarten bevorzugen diese Kunden automatische Zahlungen und haben möglicherweise auch eine geringere Kündigungsrate.

- **Elektronischer Scheck:** Kunden, die elektronische Schecks nutzen, sind oft technikaffiner und bevorzugen digitale Zahlungsmethoden.

- **Papier-Scheck:** Kunden, die Papier-Schecks nutzen, bevorzugen traditionelle Zahlungsmethoden. Diese Kunden könnten eine höhere Wahrscheinlichkeit haben, die Zahlung zu versäumen, was das Risiko einer Kündigung erhöhen kann.

- Die bevorzugte Zahlungsmethode kann wichtige Hinweise auf das Kundenverhalten und die Zufriedenheit geben. Kunden, die automatisierte und digitale Zahlungsmethoden bevorzugen, sind möglicherweise zufriedener und haben eine geringere Wahrscheinlichkeit zu kündigen.

## MonthlyCharges vs TotalCharges

Der Telco Customer Churn Datensatz enthält verschiedene Attribute, die verwendet werden, um das Kundenverhalten zu analysieren und die Wahrscheinlichkeit der Kundenabwanderung (Churn) vorherzusagen. Die Spalten "TotalCharges" und "MonthlyCharges" sind hierbei von besonderer Bedeutung, da sie direkte Einblicke in die finanziellen Interaktionen der Kunden mit dem Unternehmen bieten. Hier sind die Bedeutungen und Implikationen dieser Spalten:

**1. TotalCharges**

Die Spalte "TotalCharges" gibt den Gesamtbetrag an, den ein Kunde seit Beginn seiner Vertragslaufzeit beim Telekommunikationsanbieter gezahlt hat. Diese Variable ist kumulativ und reflektiert die gesamte finanzielle Beziehung des Kunden zum Unternehmen.

- **Langfristige Kundenbindung:** Ein höherer Gesamtbetrag kann darauf hindeuten, dass der Kunde schon länger beim Unternehmen ist und kontinuierlich dessen Dienstleistungen nutzt. Diese Kunden könnten eine stärkere Bindung zum Unternehmen haben und weniger geneigt sein, zu kündigen.

- **Kundenwert:**TotalCharges kann als Indikator für den Kundenwert verwendet werden. Kunden, die höhere Gesamtausgaben haben, könnten für das Unternehmen besonders wertvoll sein und daher besondere Aufmerksamkeit und Maßnahmen zur Kundenbindung erfordern.

- **Churn-Analyse:** Ein plötzlicher Anstieg oder Abfall in den TotalCharges kann ein Warnsignal für mögliche Unzufriedenheit und ein erhöhtes Churn-Risiko sein. Diese Informationen können verwendet werden, um proaktive Maßnahmen zu ergreifen, um die Kundenbindung zu verbessern.


**2. MonthlyCharges**

Die Spalte "MonthlyCharges" gibt an, wie viel ein Kunde monatlich für die Dienste des Telekommunikationsanbieters zahlt. Diese Variable ist ein direkter Indikator für den monatlichen finanziellen Aufwand des Kunden.

- **Preissensibilität:** Die Höhe der monatlichen Gebühren kann Aufschluss über die Preissensibilität der Kunden geben. Kunden mit höheren monatlichen Gebühren könnten eher preissensitiv sein und bei Preiserhöhungen oder vergleichbaren günstigeren Angeboten von Wettbewerbern abwandern.

- **Dienstleistungsnutzung**: MonthlyCharges können auch Aufschluss darüber geben, welche und wie viele Dienstleistungen ein Kunde nutzt. Höhere monatliche Gebühren deuten oft darauf hin, dass der Kunde mehrere oder premium Dienstleistungen abonniert hat.

- **Kundensegmente:** Durch die Analyse der monatlichen Gebühren können verschiedene Kundensegmente identifiziert werden. Beispielsweise können Kunden mit hohen monatlichen Gebühren als Premium-Kunden betrachtet werden, während Kunden mit niedrigen Gebühren als Basis-Kunden eingestuft werden können. Diese Segmente können dann gezielt mit maßgeschneiderten Angeboten und Dienstleistungen angesprochen werden.


Durch die Analyse dieser beiden Variablen kann das Unternehmen wichtige Erkenntnisse gewinnen, um gezielte Maßnahmen zur Verbesserung der Kundenbindung und zur Reduzierung der Kündigungsrate zu entwickeln. Beispielsweise können Kunden mit hohen TotalCharges und MonthlyCharges als besonders wertvoll betrachtet und mit speziellen Treueprogrammen oder Rabatten belohnt werden, um ihre Zufriedenheit und Loyalität zu erhöhen.

- **Lebenszeitwert (Customer Lifetime Value, CLV):** Die Tenure-Daten können verwendet werden, um den Lebenszeitwert der Kunden zu berechnen. Ein höherer Tenure-Wert bedeutet in der Regel einen höheren CLV, da diese Kunden länger bleiben und somit mehr Umsatz generieren.
- **Segmentierung und Zielgruppenansprache:** Durch die Analyse der Tenure-Daten können verschiedene Kundensegmente identifiziert werden. Kunden mit unterschiedlichen Bindungsdauern können unterschiedliche Bedürfnisse und Präferenzen haben. Das Unternehmen kann diese Erkenntnisse nutzen, um maßgeschneiderte Marketingkampagnen und Kundenbindungsstrategien zu entwickeln.
- **Proaktive Maßnahmen zur Kundenbindung:** Wenn Kunden mit kurzer Bindungsdauer als abwanderungsgefährdet identifiziert werden, kann das Unternehmen proaktive Maßnahmen ergreifen, um deren Zufriedenheit und Loyalität zu erhöhen. Die Abwanderungsvorhersage ist ein binäres Klassifizierungsproblem, da Kunden in einem bestimmten Zeitraum entweder abwandern oder behalten werden. Als Leitfaden für die Modellbildung müssen zwei Fragen beantwortet werden:

1- Welche Funktionen sorgen dafür, dass Kunden abwandern oder binden?

2- Was sind die wichtigsten Merkmale, um ein Modell mit hoher Leistung zu trainieren?


Lassen Sie uns die Analyse der **Information Value and Weight of Evidence** (Attributrelevanz)  verwenden, um einige Statistiken der jeweiligen Features mit dem Ziel zu erhalten.

## Information Value and Weight of Evidence

Die Analyse der Attributrelevanz bewertet den Einfluss der Attributwerte auf die Zielvariable. Sie hilft dabei, die wichtigsten Variablen zu erkennen, die den größten Einfluss auf die Zielvariable haben, und die Beziehungen und Logik zwischen den wichtigsten Prädiktoren und der Zielvariable zu verstehen.

Informationswert (IV) und Gewicht der Evidenz (WOE) sind einfache und leistungsfähige Techniken zur Durchführung der Attributrelevanzanalyse. Sie bieten einen ausgezeichneten Rahmen für explorative Analysen und werden seit Jahrzehnten umfangreich in der Welt des Kreditrisikos eingesetzt.

 **Informationswert**

 Der Informationswert bietet ein Maß dafür, wie gut eine Variable X zwischen einer binären Antwort (z. B. "gut" versus "schlecht") in irgendeiner Zielvariablen Y unterscheiden kann. Die Idee ist, dass eine Variable X, die einen niedrigen Informationswert aufweist, möglicherweise keine ausreichende Arbeit bei der Klassifizierung der Zielvariable leistet und daher als erklärende Variable entfernt wird.



**Gewicht der Evidenz**

Das Gewicht der Evidenz gibt die Vorhersagekraft einer unabhängigen Variable in Bezug auf die abhängige Variable an. Da es aus der Welt des Kredit-Scorings stammt, wird es im Allgemeinen als Maß für die Trennung von guten und schlechten Kunden beschrieben. "Schlechte Kunden" bezieht sich auf Kunden, die einen Kredit nicht zurückgezahlt haben, und "gute Kunden" bezieht sich auf Kunden, die den Kredit zurückgezahlt haben.


**Präsentation der Metriken**

Im Folgenden werden die verschiedenen in der Tabelle dargestellten Metriken erklärt, die nützlich sind, um die Beziehung zwischen der untersuchten Variable und dem Auftreten von negativen oder positiven Ergebnissen zu bewerten.

Proportion 0 oder 1 in der Antwortvariable für jeden Sektor der untersuchten Variable, was dabei hilft, die Verteilung der Variable in Bezug auf die Ergebnisse zu verstehen.


- **WoE**, eine nützliche Metrik zur Bewertung der Diskriminierung der Variable. Je weiter der WoE von 0 entfernt ist, desto diskriminierender ist die Variable. Ein negativer WoE deutet darauf hin, dass die Variable das Auftreten der Antwortvariable begünstigt, während ein positiver WoE darauf hinweist, dass die Variable das Auftreten nicht begünstigt.

- **IV**, der Informationswert, der dabei hilft, die Vorhersagekraft der Variablen zu bewerten. Der Gesamtinformationswert einer Variablen ist die Summe der IVs für jeden untersuchten Sektor. Die Berechnung des Informationswerts ermöglicht es uns, die relevantesten Variablen zu identifizieren und diese in unseren Churn-Vorhersagemodellen zu verwenden, um genauere Vorhersagen zu treffen und somit Kundenabwanderung effektiv zu bekämpfen.<br><br><br>

Die Interpretation des Informationswerts erfolgt anhand von Schwellenwerten, die festlegen, ob eine Variable als nützlich, schwach, moderat, stark oder verdächtig für die Vorhersage des Churns betrachtet wird.

**Methode und mathematische Erklärung:**



Die Berechnung des Informationswerts (IV) ist ein entscheidender Schritt in einem Projekt zur Vorhersage von Churn. Der Informationswert ist eine Metrik, die verwendet wird, um die Vorhersagekraft von Variablen zu bewerten, indem sie die Trennschärfe zwischen den verschiedenen Kategorien einer Variablen in Bezug auf die Zielvariable misst.

Um den Informationswert für eine Variable zu berechnen, führen wir in der Regel die folgenden Schritte durch:

1. **Berechnung der Häufigkeiten**: Zunächst erstellen wir eine Kreuztabelle, um die Häufigkeiten jeder Kategorie der Variable in Bezug auf die Zielvariable zu zählen. Diese Häufigkeiten sind notwendig, um die IV zu berechnen.
2. **Berechnung der Anteile**: Wir berechnen den Anteil jeder Kategorie in Bezug auf die Zielvariable. Dies geschieht, indem wir die Anzahl der Beobachtungen in jeder Kategorie durch die Gesamtanzahl der Beobachtungen teilen.
3. **Berechnung der WoE (Weight of Evidence)**: Anschließend berechnen wir das Gewicht der Evidenz (WOE) für jede Kategorie. WOE ist ein Maß dafür, wie stark eine Kategorie dazu beiträgt, das Auftreten des Ereignisses (zum Beispiel Churn) vorherzusagen. Es wird berechnet, indem der natürliche Logarithmus des Verhältnisses der Anteile der positiven und negativen Ereignisse in jeder Kategorie genommen wird.
4. **Berechnung des Informationswerts**: Schließlich berechnen wir den Informationswert für jede Kategorie, indem wir das Produkt aus (Anteil der positiven Ereignisse - Anteil der negativen Ereignisse) und WOE nehmen und über alle Kategorien summiert wird. Der Informationswert ist die Summe der Beiträge aller Kategorien einer Variable und dient als Maß für die Vorhersagekraft dieser Variable.




Die Beweiskraft (WoE)  misst die Vorhersagekraft einer unabhängigen Variablen im Verhältnis zur abhängigen Variablen.Es hat seine Wurzeln in der Welt der Kreditwürdigkeit und gibt Aufschluss über den Grad der Trennung von guten und schlechten Kunden. „Gute Kunden“ beziehen sich auf die Kunden, die einen Kredit zurückzahlen (keine Ereignisse), und „schlechte Kunden“ beziehen sich auf die Kunden, die mit der Rückzahlung eines Kredits in Verzug geraten (Ereignisse).

Die Berechnung des Informationswerts ermöglicht es uns, die relevantesten Variablen zu identifizieren und diese in unseren Churn-Vorhersagemodellen zu verwenden, um genauere Vorhersagen zu treffen und somit Kundenabwanderung effektiv zu bekämpfen.

\end{align}

IV = \sum_\ ({\%\ Non\ Event} - {\% \ Event }) *WoE
$$




   











```python
iv_list=[]
# List features
iv_features=[]
# List Klassikation
interpr_list=[]
# Dictionary sort
myDict={}

#
def iv_methode(df_clean, target):

 # List von Features außer Churn Tenure datentyp object category und int64 werden hinzufügt.
 features = df_clean.select_dtypes(include=[object,'category','int64']).columns.difference(['Churn','tenure'])
 features = list(features)

 for feature in features:

    # Berechnung von IV und WOE in eine crosstab.
    df_woe_iv = (pd.crosstab(df_clean[feature],df_clean[target],
                      normalize='columns', margins=True, margins_name='Total')
             .assign(woe=lambda dfx: np.log(dfx['No'] / dfx['Yes']))
             .assign(iv=lambda dfx: (dfx['woe']*
                                           (dfx['No']-dfx['Yes']))))

    # List mit Features Ergbnisse von IV und WOE
    iv_features.append(feature)

    # Summe von IV für jede features
    iv_list.append(round(df_woe_iv['iv'].sum(),6))

    #new_item = {'feature':feature, 'iv_sum':df_woe_iv['iv'].sum()}
    #myDict.update({'feature':feature, 'iv_sum':df_woe_iv['iv'].sum()})

    #  Dictionary mit Features und Summe.
    myDict['features'] = iv_features
    myDict['sum_iv'] = iv_list

    # Ausgabe von IV Summe
    display('IV Summe {}: {}'.format(feature,round(df_woe_iv['iv'].sum(),6)))
    #display("------------------------------------")
    print(' ')
    print(' ')
    display(df_woe_iv.style.format('{:.5}'))

    print(' ')
    print(' ')

 return  myDict;

# Funktion für Klassifizierung von den IV Ergebnisse.
def iv_interpretation(myDict):

  for iv in myDict['sum_iv']:

      if iv < 0.02:
        iv_interpretation = 'useless'

      elif iv < 0.1:
        iv_interpretation = 'weak'

      elif iv < 0.3:
          iv_interpretation = 'medium'

      elif iv < 0.5:
          iv_interpretation = 'strong'

      else:
          iv_interpretation = 'suspicious'

      # List mit der Klassifizierung
      interpr_list.append(iv_interpretation)

  #  Dictionary mit interpretation, Dictionary wird später sortiert.
  myDict['interpretation'] = interpr_list

  return myDict;


# Funktion sortieren
def sort_interpretation(myDict):

  # Dictionary nach 'sum_iv' sortieren
  sorted_myDict = sorted(zip(myDict['sum_iv'], myDict['features'], myDict['interpretation']), reverse=True)

  #  Werte sortieren nach  Features, IV und Interpretation.
  sorted_sum_iv = [item[0] for item in sorted_myDict]
  sorted_features = [item[1] for item in sorted_myDict]
  sorted_interpretation = [item[2] for item in sorted_myDict]

  # Sortierte Dictionary erstellen
  sorted_dict = {'features': sorted_features, 'sum_iv': sorted_sum_iv, 'interpretation': sorted_interpretation}

  # Ausgabe des sortierten Dictionarys
  # print(sorted_dict)

  return  sorted_dict;

# Funktion für die sortiert Ausgabe von Values und Features
def table(sorted_dict):
 resultList_key = list(sorted_dict.keys())
 resultList_values = list(sorted_dict.values())

 #print(resultList_key)
 #print(resultList_values)


# Table von Resultat
 table = go.Table(
    header=dict(values=resultList_key ),
    cells=dict(values=resultList_values)
     )

# layout bearbeitung
 fig = go.Figure(data=table).update_layout(title="IV Summe",
 autosize=True,
 width=1010,
 height=650,)


 fig.show()
```

```python
# IV Methode ausführen.
iv = iv_methode(df_category,'Churn')

# IV interpretieren.
dic = iv_interpretation(iv)

# Sortieren von IV
sorted_dict = sort_interpretation(dic)

#table(sorted_dict)
```

Recall ist eine wichtige Metrik, insbesondere in Szenarien, in denen es kritisch ist, möglichst viele positive Fälle (in diesem Fall, churnende Kunden) zu identifizieren. Beim Telco Customer Churn Datensatz ist die Verwendung von Recall aus mehreren Gründen besonders sinnvoll:

**Bedeutung von Recall bei Customer Churn**

**1- Kundenabwanderung erkennen:**

- Hohe Kosten der Abwanderung: Kundenabwanderung kann für ein Unternehmen erhebliche finanzielle Verluste bedeuten. Es ist daher entscheidend, so viele churnende Kunden wie möglich zu identifizieren, um gezielte Maßnahmen ergreifen zu können, die diese Kunden halten könnten.

Hierfür ist es wichtiger, möglichst viele churnende Kunden korrekt zu identifizieren (hoher Recall), selbst wenn dies auf Kosten einer höheren Anzahl von False Positives geht.


**2- Konsequenzen von False Negatives:**


- Verpasste Chancen: Wenn ein Modell churnende Kunden nicht erkennt (False Negatives), verpasst das Unternehmen die Gelegenheit, diese Kunden zu halten. - Kundenbeziehungen: Bei der Kundenbindung ist es besser, mehr potenzielle Abwanderer anzusprechen, auch wenn einige davon tatsächlich nicht abwandern werden, als churnende Kunden unbeachtet zu lassen.

<br>

Recall ist bei der Vorhersage von Kundenabwanderung besonders wichtig, da das primäre Ziel darin besteht, möglichst viele churnende Kunden zu identifizieren, um gezielte Maßnahmen zur Kundenbindung ergreifen zu können. Ein hoher Recall minimiert die Anzahl der übersehenen churnenden Kunden und maximiert die Effektivität von Kundenbindungsstrategien.

Irei usar a cross-validation para estimar o erro da baseline e dos modelos iniciais. Para facilitar e melhorar o código, foi criada uma função `val_model`. A métrica utilizada para avaliação será `recall`.

```python
def val_model(X, y, clf, quite=False):
    """
    Realiza cross-validation com os dados de treino para determinado modelo.

    # Arguments
        X: DataFrame, contém as variáveis independetes.
        y: Series, vetor contendo a variável alvo.
        clf: modelo classificador do Scikit-learn.
        quite: bool, indicando se a função deve imprimir os resultados ou não.

    # Returns
        float, média dos scores da cross-validation.
    """

    X = np.array(X)
    y = np.array(y)

    pipeline = make_pipeline(StandardScaler(), clf)
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    if quite == False:
        print("Recall: {:.2f} (+/- {:.2f})".format(scores.mean(), scores.std()))

    return scores.mean()
```

```python
# criar baseline e ver desempenho
rf = RandomForestClassifier()
score_baseline = val_model(X_train, y_train, rf)
```

A técnica escolhida foi Under Sampling, após alguns teste de desempenho das outras. Seguindo recomendações de algumas literaturas, os dados serão padronizados antes de usar essa técnica de balanceamento.

## Bedeutung verschiedener Features

Ein Generalized Linear Model (GLM) wird oft verwendet, um die Bedeutung verschiedener Features in einem Datensatz zu bewerten, insbesondere wenn die Zielvariable binär ist, wie es bei der Vorhersage von Kundenabwanderung (Churn) der Fall ist. Hier sind die Gründe und Schritte, warum und wie ein GLM verwendet wird:

**Gründe für die Verwendung eines GLM:**

**1-Flexibilität:** GLMs sind flexibel und können verschiedene Verteilungen der Zielvariablen modellieren. Bei binären Zielvariablen (wie Churn: Ja/Nein) wird häufig die Binomialverteilung verwendet.

**2-Interpretierbarkeit:** Die Koeffizienten eines GLMs sind relativ leicht zu interpretieren. Sie geben an, wie sich Änderungen in den Features auf die Wahrscheinlichkeit der Zielvariable auswirken.

**3-Bewertung der Signifikanz:** GLMs bieten statistische Tests (z.B. p-Werte) zur Bewertung der Signifikanz jedes Features. Dies hilft dabei zu entscheiden, welche Features beibehalten werden sollten.

**4-Standardisierte Methode:** GLMs sind gut erforscht und standardisiert, was sie zu einer vertrauenswürdigen Methode für die Bewertung der Relevanz von Features macht.

<br>

**Interpretation der Ergebnisse:**

**1- Koeffizienten:** Die Koeffizienten geben die Stärke und Richtung des Einflusses jedes Features auf die Zielvariable an.

**2- p-Werte:** Niedrige p-Werte (typischerweise < 0.05) deuten darauf hin, dass das Feature signifikant ist und einen Einfluss auf die Zielvariable hat.

**3- Odds Ratios:** Können berechnet werden, um die Wahrscheinlichkeit zu interpretieren.

<br>
<br>

Die Abwanderungsvorhersage ist ein binäres Klassifizierungsproblem, da Kunden in einem bestimmten Zeitraum entweder abwandern oder bleiben. Zur Modellerstellung müssen zwei Fragen beantwortet werden:

**1- Welche Merkmale führen zu Kundenabwanderung oder Kundenbindung?**

**2- Was sind die wichtigsten Merkmale, um ein Modell mit hoher Leistung zu trainieren?**

wir verwenden das verallgemeinerte lineare Modell (GLM), um einige Statistiken der jeweiligen Merkmale mit dem Ziel zu erhalten.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Ändere die Trennzeichen für Variablennamen in '_'
# Hier werden alle Spaltennamen im DataFrame df_proc so verändert, dass Leerzeichen, Klammern und Bindestriche durch Unterstriche (_) ersetzt werden.
# Dies ist notwendig, um sie in einer Formel für das Generalisierte Lineare Modell (GLM) verwenden zu können, da diese Zeichen in Formeln nicht erlaubt sind.
all_columns = [column.replace(" ", "_").replace("(", "_").replace(")", "_").replace("-", "_") for column in df_proc.columns]

# Führe die Änderung an den Spaltennamen des Dataframes durch
df_proc.columns = all_columns

# Führe die Änderung an den Spaltennamen des Dataframes durch
glm_columns = [e for e in all_columns if e not in ['customerID', 'Churn']]

# Dann werden die verbleibenden Spaltennamen zu einem einzigen String zusammengefügt, wobei jeder Name durch + getrennt wird. Dieser String wird später in der Formel verwendet.
glm_columns = ' + '.join(map(str, glm_columns))

# 'formula=f'Churn ~ {glm_columns}' erstellt eine Formel, bei der Churn die Zielvariable ist und alle anderen vorbereiteten Spalten als Prädiktoren dienen.
glm_model = smf.glm(formula=f'Churn ~ {glm_columns}', data=df_proc, family=sm.families.Binomial())

# An das verallgemeinerte lineare Modell anpassen
res = glm_model.fit()
print(res.summary())
```

```python
# Extrahieren die signifikante Spalten basierend auf p-Werten
significant_columns = [col for col in res.pvalues.index if res.pvalues[col] < 0.05 and col != 'Intercept']

for significant_column in significant_columns:
    print(significant_column)

```

**Für die erste Frage sollten Sie sich die Spalte (P>|z|) ansehen. Wenn der absolute p-Wert kleiner als 0,05 ist, bedeutet dies, dass das Feature die Abwanderung statistisch signifikant beeinflusst. Beispiele sind:**

- InternetService_Fiber_optic[T.True]
- InternetService_No[T.True]
- Contract_Month_to_month[T.True]
- Contract_Two_year[T.True]
- PaymentMethod_Electronic_check[T.True]
- SeniorCitizen
- tenure
- MultipleLines
- PaperlessBilling
- TotalCharges

**Die zweite Frage zu den Feature-Wichtigkeiten kann durch Betrachtung der Exponentialkoeffizientenwerte beantwortet werden. Der Exponentialkoeffizient schätzt die erwartete Änderung der Fluktuation durch ein bestimmtes Feature bei einer Änderung von einer Einheit.**
